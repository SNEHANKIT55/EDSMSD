```python
#matrix multiplication using NumPy

import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

c = np.dot(a,b)

print(c)
```

```python
# stack two arrays vertically using NumPy
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

c = np.vstack((a,b))

print(c)
```

```python
#generate a custom sequence using NumPy

import numpy as np

a = np.arange(0, 10, 2)

print(a)
```

```python
#arithmetic operations on NumPy arrays
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

c = a + b

print(c)
```

```python
# bitwise operations on NumPy arrays
import numpy as np

a = np.array([1, 2], dtype=np.uint8)
b = np.array([4, 5], dtype=np.uint8)

c = np.bitwise_and(a,b)

print(c)
```

```python
# stack two arrays horizontally using NumPy:
import numpy as np

a = np.array([1, 2], dtype=np.uint8)

b = a.copy()

c = a.view()

print(b)
print(c)
```

```python
#search for elements in a NumPy array
import numpy as np

a = np.array([1, 2])

b = (a == 2)

print(b)
```

```python
#count elements in a NumPy array
import numpy as np

a = np.array([3, 1, 2])


b = np.sort(a)

print(b)
```

```python
#broadcast elements in a NumPy array
import numpy as np

a = np.array([[1], [2]])
b = np.array([3])

c = a + b

print(c)
```
